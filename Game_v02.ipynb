```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# number of unique items
num_of_unique_items = len(purchase_data['Item ID'].unique())
avg_price = purchase_data["Price"].mean()
num_of_purchase = len(purchase_data['Purchase ID'].unique())
total_rev = purchase_data["Price"].sum()

purchase_summary = pd.DataFrame({"Number of Unique Items":num_of_unique_items,
                                            "Average Purchase Price":avg_price,
                                            "Total Num of Purchase":num_of_purchase,
                                            "Total Revenue":total_rev
},index=[0])

purchase_summary.head()
```

## Gender Demographics

```python
# pull in the df with only unique players
# count male players and female players
gender_counts = player_info_df["Gender"].value_counts()

# count % for each gender
male_per = gender_counts["Male"]/gender_counts.sum()
female_per = gender_counts["Female"]/gender_counts.sum()
other_per = 1-male_per-female_per

# create summary df
gender_summary = pd.DataFrame({"Total Count":gender_counts,"Percentage of Players":[male_per,female_per,other_per]})
gender_summary.reset_index()
gender_summary.head()

# why couldn't I reset index to be 0,1,2??
```

```python
purchase_data.head()
group_by_sex = purchase_data.groupby("Gender")
c1 = group_by_sex["Purchase ID"].count()
c2 = group_by_sex["Price"].mean()
c3 = group_by_sex["Price"].sum()

gender_purchase_summary = pd.DataFrame({"Purchase Count":c1,"Avg Purchase Price":c2,"Total Purchase Value":c3})
gender_purchase_summary=pd.merge(gender_purchase_summary,gender_summary,left_index=True,right_index=True)
gender_purchase_summary.head()
gender_purchase_summary["Avg Total Purchase per Person"]=gender_purchase_summary["Total Purchase Value"]/gender_purchase_summary["Total Count"]
del gender_purchase_summary["Percentage of Players"]
del gender_purchase_summary["Total Count"]
gender_purchase_summary.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# saved output. no code
```

```python
bins=[0,9,14,19,24,29,34,39,200]
labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
player_info_df["Age Group"]=pd.cut(player_info_df["Age"],bins,labels=labels)
player_info_df.head()
```

```python
grouped = player_info_df.groupby("Age Group")
count_by_age_group = grouped["SN"].count()
percentage_by_age_group = grouped["SN"].count()/num_of_unique_players
age_gp_summary = pd.DataFrame({"Total Count":count_by_age_group,"Percentage of Players":percentage_by_age_group})
age_gp_summary.head(20)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# keep the output to match it
```

```python
purchase_data["Age Group"]=pd.cut(purchase_data["Age"],bins,labels=labels)
purchase_data.head()
```

```python
purchase_grouped = purchase_data.groupby("Age Group")
c1 = purchase_grouped["Purchase ID"].count()
c2 = purchase_grouped["Price"].mean()
c3 = purchase_grouped["Price"].sum()
age_gp_purchase_summary = pd.DataFrame({"Purchase Count":c1,"Average Purchase Price":c2,"Total Purchase Value":c3})
pd.options.display.float_format = '{:.2f}'.format
age_gp_purchase_summary = pd.merge(age_gp_summary,age_gp_purchase_summary,left_index=True,right_index=True)
age_gp_purchase_summary=age_gp_purchase_summary.rename(columns={"Total Count":"Total # of players"})
age_gp_purchase_summary["Avg Total Purchase per Person"] = age_gp_purchase_summary["Total Purchase Value"]/age_gp_purchase_summary["Total # of players"]
del age_gp_purchase_summary["Total # of players"]
del age_gp_purchase_summary["Percentage of Players"]
age_gp_purchase_summary.head()
```
